```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 'top1', 74.5520000415039
with open("train_imagenet.log") as f:
    data = f.read()
top1 = np.array([float(num) for num in re.findall(r"top1', (\d+\.\d+)", data)][:88])

with open("../train_imagenet.log") as f:
    data = f.read()
top1_undo = np.array([float(num) for num in re.findall(r"top1', (\d+\.\d+)", data)][:90])
top1_undo[:10] = top1[:10]
top1 = np.concatenate([top1, top1_undo[88:]])
print(top1_undo)

df = pd.DataFrame(top1, columns=["top1"])
ax = df.plot()

ax.plot(top1_undo, label="top1 with undo")
```
