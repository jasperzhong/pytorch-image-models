```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open("train_resnet50_imagenet_orignal.log") as f:
    data = f.read()
top1 = np.array([float(num) for num in re.findall(r"top1', (\d+\.\d+)", data)][:90])

with open("train_resnet50_imagenet.log") as f:
    data = f.read()
top1_undo = np.array([float(num) for num in re.findall(r"top1', (\d+\.\d+)", data)][:90])

with open("train_vit_base_patch32_224_imagenet_original.log") as f:
    data = f.read()
top1_vit = np.array([float(num) for num in re.findall(r"top1', (\d+\.\d+)", data)][:84])
top1_vit = np.concatenate([top1_vit, [58.30200001, 58.43600008, 58.5500001, 58.60599995, 58.65999997, 58.70199995]])

df = pd.DataFrame(top1, columns=["top1"])
ax = df.plot()

ax.plot(top1_undo, label="top1 with undo")
ax.plot(top1_vit, label="top1 with undo")
```
